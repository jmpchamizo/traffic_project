## Comenzamos cargando los datos y viendo su estructura

```python
import pickle, random
import matplotlib.pyplot as plt
```

```python
infile = open('../data/data8.pickle','rb')
new_dict = pickle.load(infile)
new_dict.keys()
```

### En labels tenemos la señal a la que corresponde cada código:

```python
print(new_dict['labels'][0])
print(new_dict['labels'][14])
```

### Veamos los datos con los que entrenaremos:

```python
y_train = new_dict['y_train']
y_test = new_dict['y_test']
x_train = new_dict['x_train'][:,0,:,:]
x_test = new_dict['x_test'][:,0,:,:]
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
plt.hist(y_train, len(set(y_train)));
```

### Tenemos 2000 fotos por label(señal)

### Vamos a ver que señales tenemos y como son las señales:

```python
plt.figure(figsize=(20,20))
for i,label in enumerate(new_dict['labels']):
    ind=-1
    for j,_ in enumerate(y_train):
        if y_train[j] == i:
            ind = j
            break
    plt.subplot(7,7, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.title(label)
    plt.ylabel(str(x_train[ind].shape))
    plt.imshow(x_train[ind])
```
