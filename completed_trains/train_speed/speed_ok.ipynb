### Creamos pos.info global

```python
fout = open("result.txt", "wt")
inf = glob.glob('pos*.txt')
for i in inf:
    fin = open(i,"rt")
    for line in fin:
        fout.write(f'{line.split(".")[0]}_{i.split(".")[0][-2:]}.{line.split(".")[1]}')
    fin.close()
fout.close()
```
