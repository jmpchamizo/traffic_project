```python
import cv2
import pickle
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import os
import glob
```

## Tratamos im√°genes negativas

```python
archivos = glob.glob("train/negatives/*.jpg")
```

```python
imagenes = [cv2.imread(archivo) for archivo in archivos]
len(imagenes)
```

```python
imagenes_ch = [imagen[:,:,0] for imagen in imagenes]

imagenes_res = [cv2.resize(imagen, (32,32), interpolation = cv2.INTER_AREA) for imagen in imagenes_ch]
```

```python
os.remove("bg.txt")
f= open("bg.txt","w+")
for i in range(500):
    matplotlib.image.imsave(f'data/training/stop/neg/neg{i}.jpg', imagenes[i])
    f.write(f"data/training/stop/neg/neg{i}.jpg\n" )
```

## Positivas 

```python
infile = open('data/data8.pickle','rb')
new_dict = pickle.load(infile)
stops = new_dict["x_train"][new_dict['y_train']==14]
```

```python
os.remove("stop.info")
f= open("stop.info","w+")
for i in range(500):
    matplotlib.image.imsave(f'data/training/stop/pos/stop{i}.jpg', stops[i,0,:,:])   
    f.write(f"data/training/stop/pos/stop{i}.jpg 1 0 0 32 32\n" )
f.close()
```
